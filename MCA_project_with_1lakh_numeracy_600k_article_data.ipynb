```python
def convert(text,off,l,n):
  text1=text[:off]+'[NUM]'+text[off+l+1:]
  b=len(text1)
  a=text1.find('[NUM]')+5
  if a<b:
    if text1[a]!=' ':
      text1=text1[:a]+' '+text1[a:]
  p=text1.find('[NUM]')-1
  if p>0:
    if text1[p]!=' ':
      text1=text1[:p]+' '+text1[p:]
  return text1
```

```python
corpus1=[]
for i in range(0,len(corpus)):
    text=convert(corpus[i],corpusoff[i],corpuslen[i],len(corpus[i]))
    corpus1.append(text)
```

```python
from transformers import AutoTokenizer
import torch
from transformers import BertTokenizer
from transformers import BertModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("nlpaueb/sec-bert-num")
model = BertModel.from_pretrained('nlpaueb/sec-bert-num')
```

```python
def tokenizetexttensor(text):
  tokenized_text = tokenizer.tokenize(text)
  indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
  index = tokenized_text.index('[NUM]')
  tokens_tensor = torch.tensor([indexed_tokens])
  with torch.no_grad():
    last_hidden_states = model(tokens_tensor)[0] 
  embedding_of_num = last_hidden_states[:,index,:]
  embedding_of_num_use = embedding_of_num[0].cpu().detach().numpy()
  return embedding_of_num_use
```

```python
def tensorout(list):
  i=0
  output=[]
  for x in list:
    print(i)
    i=i+1
    out=tokenizetexttensor(x)
    output.append(out)
  with open('/content/drive/MyDrive/Exnum/test_txt3.txt','w+') as f:
    np.savetxt(f,np.array(output))
  f.close()
  return output
```

```python
from sklearn.neural_network import MLPClassifier
import optuna
def objective(trial):
  n_layers=trial.suggest_int('n_layers',1,4)
  layers=[]
  for i in range(n_layers):
    layers.append(trial.suggest_int(f'n_units_{i}',50,300))
  clf5 = MLPClassifier(hidden_layer_sizes=tuple(layers),
                    random_state=5,
                    verbose=True,
                    learning_rate_init=0.01,
                    max_iter=30
                    )
  clf5.out_activation_='softmax'
  clf5.fit(x_train_feature,y_train_feature)
  return clf5.score(x_test_feature,y_test_feature)
study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=5)
```

```python
trial=study.best_trial
print('accuracy : {}'.format(trial.values))
print("best parameter : {}".format(trial.params))
```
