```python
from transformers import AutoTokenizer
import torch
from transformers import BertTokenizer
from transformers import BertModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("nlpaueb/sec-bert-num")
model = BertModel.from_pretrained('nlpaueb/sec-bert-num')
```

```python
def tokenizetexttensor(text):
  tokenized_text = tokenizer.tokenize(text)
  indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
  index = tokenized_text.index('[NUM]')
  tokens_tensor = torch.tensor([indexed_tokens])
  with torch.no_grad():
    last_hidden_states = model(tokens_tensor)[0] 
  embedding_of_num = last_hidden_states[:,index,:]
  embedding_of_num_use = embedding_of_num[0].cpu().detach().numpy()
  return embedding_of_num_use
```

```python
def tensorout(list,path):
  i=0
  output=[]
  for x in list:
    print(i)
    i=i+1
    out=tokenizetexttensor(x)
    output.append(out)
  with open(path,'w+') as f:
    np.savetxt(f,np.array(output))
  f.close()
  return output
```

```python
import optuna
```

```python
def objective(trial):
  n_estimators=trial.suggest_int('n_estimators',50,500)
  max_depth=trial.suggest_int('max_depth',1,15)
  max_leaf_nodes=trial.suggest_int('max_leaf_nodes',50,500)
  min_samples_split=trial.suggest_int('min_samples_split',50,500)
  hcls=RandomForestClassifier(n_estimators=n_estimators,max_depth=max_depth,max_leaf_nodes=max_leaf_nodes,min_samples_split=min_samples_split,n_jobs=-1)
  hcls.fit(x_train_feature,y_train)
  return hcls.score(x_test_feature,y_test)
study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=100)
```

```python
trial=study.best_trial
print('accuracy : {}'.format(trial.values))
print("best parameter : {}".format(trial.params))
```

```python
def objective(trial):
  solver=trial.suggest_categorical('solver',['newton-cg','lbfgs','liblinear'])
  C=trial.suggest_float('C',0.01,100)
  penalty='l2'
  max_iter=1000
  hcls1=LogisticRegression(solver=solver,max_iter=max_iter,penalty=penalty,C=C,n_jobs=-1)
  hcls1.fit(x_train_feature,y_train)
  return hcls1.score(x_test_feature,y_test)
study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=100)
```
