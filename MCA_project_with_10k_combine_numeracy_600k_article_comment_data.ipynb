```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, GlobalMaxPooling1D, Embedding, Conv1D ,Dropout
from tensorflow.python.keras import regularizers
```

```python
cnn_model = Sequential()

embedding_layer = Embedding(9000,768, weights=[x_train], input_length=768 , trainable=False)
cnn_model.add(embedding_layer)

cnn_model.add(Conv1D(500, 7, activation='relu',padding='same'))
cnn_model.add(Conv1D(400, 5, activation='relu',padding='same'))
cnn_model.add(Conv1D(300, 4, activation='relu',padding='same'))
cnn_model.add(GlobalMaxPooling1D())
cnn_model.add(Dense(17, activation='relu'))
cnn_model.add(Dense(7, activation='softmax'))
```

```python
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
print(cnn_model.summary())
```

```python
cnn_model_history = cnn_model.fit(x_train, y_train, batch_size=128, epochs=20, verbose=1, validation_split=0.2)
```

```python
cnn_model.evaluate(x_test,y_test)
```

```python
y_train_pred=cnn_model.predict(x_train)
```

```python
y_test_pred=cnn_model.predict(x_test)
```

**Lstm**

```python
from keras.layers import LSTM
lstm_model = Sequential()
embedding_layer = Embedding(9000,768, weights=[x_train], input_length=768 , trainable=False)

lstm_model.add(embedding_layer)
lstm_model.add(LSTM(300,return_sequences=True))
lstm_model.add(LSTM(200,return_sequences=True))
lstm_model.add(LSTM(100))
lstm_model.add(Dense(17, activation='relu'))
lstm_model.add(Dense(7, activation='softmax'))
```

```python
lstm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
print(lstm_model.summary())
```

```python
lstm_model_history = lstm_model.fit(x_train, y_train, batch_size=128,epochs=50, verbose=1, validation_split=0.2)
```
